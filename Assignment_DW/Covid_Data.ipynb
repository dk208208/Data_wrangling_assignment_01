### 4. Filter the dataframe with only this columns
   ['continent','location','date','total_cases','total_deaths','gdp_per_capita','
   human_development_index'] and update the data frame.
 

```python
dataset = dataset[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
dataset.head()
```

### 5. Data Cleaning
        a. Remove all duplicates observations
        b. Find missing values in all columns
        c. Remove all observations where continent column value is missing
        Tip : using subset parameter in dropna
        d. Fill all missing values with 0
 

#### a. Remove all duplicates observations

```python
dataset.drop_duplicates()
print("No. of Duplicates : ",dataset.duplicated().sum())
```

#### b. Find missing values in all columns

```python
dataset.isnull().sum()
```

#### c. Remove all observations where continent column value is missing. 

```python
dataset.dropna(subset = ["continent"])
```

#### d. Fill all missing values with 0. 

```python
dataset = dataset.fillna(0)
```

```python
dataset.isna().sum()
```

### 6. Date time format :
        a. Convert date column in datetime format using pandas.to_datetime
        b. Create new column month after extracting month data from date
        column.


#### a. Convert date column in datetime format using pandas.to_datetime

```python
dataset['date'] = pd.to_datetime(dataset['date'])
```

#### b. Create new column month after extracting month data from date column.

```python
dataset["month"] = dataset["date"].dt.month
dataset
```

### 7. Data Aggregation:
        a. Find max value in all columns using groupby function on 'continent'
        column
        Tip: use reset_index() after applying groupby
        b. Store the result in a new dataframe named 'df_groupby'.
        (Use df_groupby dataframe for all further analysis)


#### a. Find max value in all columns using groupby function on 'continent' column.

```python
df = dataset.groupby(by ="continent").max()
df
```

#### b. Store the result in a new dataframe named 'df_groupby'.

```python
df_groupby = pd.DataFrame(df)
df_groupby
```

### 8. Feature Engineering :
        a. Create a new feature 'total_deaths_to_total_cases' by ratio of
        'total_deaths' column to 'total_cases'


```python
dataset['total_deaths_to_total_cases'] = dataset['total_deaths']/dataset['total_cases']
dataset.head(10)
```

### 9. Data Visualization :
        a. Perform Univariate analysis on 'gdp_per_capita' column by plotting
        histogram using seaborn dist plot.
        b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
        c. Plot Pairplot on df_groupby dataset.
        d. Plot a bar plot of 'continent' column with 'total_cases' .


#### a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.

```python
sns.distplot(a=dataset['gdp_per_capita'],color='blue')
```

#### b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'.

```python
sns.scatterplot(x=dataset['total_cases'],y=dataset['gdp_per_capita'],hue=dataset['total_deaths'])
```

#### c. Plot Pairplot on df_groupby dataset.

```python
sns.pairplot(data = df_groupby)
```

#### d. Plot a bar plot of 'continent' column with 'total_cases' .

```python
sns.catplot(data=dataset, x='continent', y='total_cases',kind='bar')
```
